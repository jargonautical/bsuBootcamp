# Dealing with text data once you have it

## Beautiful Soup

```python
site = "https://jargonautical.github.io/open-data-aqi/"
```

```python
import requests
from bs4 import BeautifulSoup # documentation at https://beautiful-soup-4.readthedocs.io/en/latest/#
s = requests.get(site)
soup = BeautifulSoup(s.content, 'html.parser')
for d in soup.find_all('div'):
    print(d.get_text)
```
