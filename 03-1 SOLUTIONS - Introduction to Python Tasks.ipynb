3.a) Create another function that will check if the user’s answer matches the actual answer.

````python
import random

def addition():
    num1 = random.randint(5, 20)
    num2 = random.randint(5, 20)
    user_answer = float(input(f"What is {num1} + {num2}? "))
    correct_answer = num1 + num2
    return user_answer, correct_answer

def subtraction():
    num1 = random.randint(25, 50)
    num2 = random.randint(1, 25)
    user_answer = float(input(f"What is {num1} - {num2}? "))
    correct_answer = num1 - num2
    return user_answer, correct_answer

def check_answer(user_answer, correct_answer):
    if user_answer == correct_answer:
        print("Correct")
    else:
        print(f"Incorrect, the answer is {correct_answer}")

def main():
    while True:
        print("1. Addition")
        print("2. Subtraction")
        choice = input("Enter 1 or 2: ")

        if choice == '1':
            user_answer, correct_answer = addition()
            check_answer(user_answer, correct_answer)
            break
        elif choice == '2':
            user_answer, correct_answer = subtraction()
            check_answer(user_answer, correct_answer)
            break
        else:
            print("Invalid selection. Please enter 1 or 2.")

if __name__ == "__main__":
    main()
````

4.a OPTIONAL TASK

In Python, it is not technically possible to directly delete a record from a .csv file.


Instead you need to save the file to a temporary list in Python, make the changes to the list and then overwrite the original file with the temporary list.


Change the previous program to allow you to do this.

Your menu should now look like this:

```
1) Add to file
2) View all records
3) Delete a record
4) Quit Program

Enter the number of your selection:
```

````python
import csv

def add_to_file():
    name = input("Enter your name: ")
    salary = input("Enter your salary: ")
    with open('Salaries.csv', 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([name, salary])

def view_all_records():
    with open('Salaries.csv', 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            print(row)

def delete_a_record():
    name_to_delete = input("Enter the name of the record to delete: ")
    records = []
    with open('Salaries.csv', 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if row[0] != name_to_delete:
                records.append(row)
    with open('Salaries.csv', 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(records)

def main():
    while True:
        print("1) Add to file")
        print("2) View all records")
        print("3) Delete a record")
        print("4) Quit program")
        choice = input("Enter the number of your selection: ")

        if choice == '1':
            add_to_file()
        elif choice == '2':
            view_all_records()
        elif choice == '3':
            delete_a_record()
        elif choice == '4':
            break
        else:
            print("Invalid selection. Please enter 1, 2, 3 or 4.")

if __name__ == "__main__":
    main()
````

6. Make a maths quiz that asks five questions by randomly generating two whole numbers to make the question (e.g. [num1] + [num2]).

If they get it right add a point to their score.
At the end of the quiz, tell them how many they got correct out of five.

````python
import random

# Initialize the score
score = 0

# Ask five questions
for i in range(1,6):
    # Randomly generate two whole numbers
    num1 = random.randint(1, 50)
    num2 = random.randint(1, 50)

    # Ask the user to enter the answer
    answer = int(input(f"What is {num1} + {num2}? "))

    # If they get it right add a point to their score
    if answer == num1 + num2:
        score += 1

# At the end of the quiz, tell them how many they got correct out of five
print(f"You got {score} correct out of 5.")
````

7. Display five colours and ask the user to pick one.
If they pick the same as the program has chosen, say “Well done”, otherwise display a witty answer which involves the correct colour, e.g. “I bet you are GREEN with envy” or “You are probably feeling BLUE right now”.
Ask them to guess again; if they have still not got it right, keep giving them the same clue and ask the user to enter a colour until they guess it correctly.

````python
import random

# Define a list of five colours
colours = ["red", "blue", "green", "yellow", "purple"]

# Randomly choose a colour
chosen_colour = random.choice(colours)

# Define witty responses for each colour
witty_responses = {
    "red": "Seeing red now, are we?",
    "blue": "Feeling a bit blue, aren't you?",
    "green": "I bet you are green with envy.",
    "yellow": "Not quite the yellow brick road, is it?",
    "purple": "No need to be purple with rage."
}

while True:
    # Ask the user to pick a colour
    user_choice = input("Pick a colour (red, blue, green, yellow, purple): ")

    # Check if the user's choice matches the chosen colour
    if user_choice == chosen_colour:
        print("Well done!")
        break
    else:
        # Display a witty answer which involves the correct colour
        print(witty_responses[chosen_colour])
````
